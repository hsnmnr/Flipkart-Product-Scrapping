```python
import requests

def mapquest(loc):
    key = 'X6PzGCMX5gL6OckukuR8kwfoKoEJeBvO'
    url = 'http://open.mapquestapi.com/geocoding/v1/address?outFormat=csv&key='+key+'&location='+loc
    results = requests.get(url)
    for line in results.text.splitlines():
        data = line.split(",")
        print(data)
        lat  = data[6]
        lon  = data[7]

        lat  = lat.replace('"','')
        lon  = lon.replace('"','')
#         print(lat)
#         print(lon)
    return lat,lon
mapquest('pakistan')
```

```python
class UFO(object):        
    def __init__(self, city,state, count):
        self.city = city
        self.state = state
        self.count = count
```

```python
urls  = ['http://www.cs.umd.edu/~golbeck/INST326/ndxe201902.html',
         'http://www.cs.umd.edu/~golbeck/INST326/ndxe201903.html' ]
```

```python
from bs4 import BeautifulSoup as BeautifulSoup  # HTML data structure
from urllib.request import urlopen as uReq  # Web client
```

```python
cities = []
states = []
for url in urls:
    page = requests.get(url)
    soup = BeautifulSoup(page.text,'html.parser')
    rows = soup.find_all("tr")
    c=0 
    for tr in rows:
        cols = tr.findAll('td')
        for td in cols:
            c+=1
            if(c==2): # second column is city
                cities.append(td.get_text())
            if(c==3):  # third column is state
                states.append(td.get_text())
        c=0    
```

```python
print(states)
```

```python
cities = list(filter(lambda x: x != "", cities) ) # removing empty strings
states = list(filter(lambda x: x != "", states) )
```

```python
print( "First 5 cities = ",cities[:5])
print( "Total = ",len(cities), " Unique = ", len(list(set(cities))) )
```

```python
unique_cities = {}
for key,value in zip(cities,states): 
    if value not in unique_cities.keys():  # find unique cities and join with states
        unique_cities[key] = value
```

```python
print(unique_cities)
```

```python
objects = []

cities_counts = {}
for city,state in unique_cities.items():
    counts = cities.count(city)  # number of times a city appeared 
    objects.append(UFO(city,state,counts)) # create object for every city
```
